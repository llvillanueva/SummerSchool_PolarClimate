```python
#modules
import xarray as xr
import numpy as np
import cftime
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
#file path
path = "/afs/ictp.it/home/t/tkyeimia/for_students/Data/"

#variables
var_pr = "pr"
var_prc = "prc"
var_qflx = "qflx"

#experiments
exp1 = "pa-pdSIC-ext"
exp2 = "pa-futArcSIC-ext"
exp3 = "pa-futAntSIC-ext"
exp4 = "pa-pdSIC-2XCO2-ext"
exp5 = "pa-futSIC-2XCO2-ext"
```

```python
#openning datasets

#precipitation
pr_pdSIC=xr.open_dataset(path+var_pr+"_Amon_CESM1-WACCM-SC_"+exp1+"_combined.nc")[var_pr]
pr_futArcSIC=xr.open_dataset(path+var_pr+"_Amon_CESM1-WACCM-SC_"+exp2+"_combined.nc")[var_pr]
pr_pdSIC_2XCO2=xr.open_dataset(path+var_pr+"_Amon_CESM1-WACCM-SC_"+exp4+"_combined.nc")[var_pr]
pr_futSIC_2XCO2=xr.open_dataset(path+var_pr+"_Amon_CESM1-WACCM-SC_"+exp5+"_combined.nc")[var_pr]

#evaporation - water flux
qflx_pdSIC=xr.open_dataset(path+var_qflx+"_Amon_CESM1-WACCM-SC_"+exp1+"_combined.nc")[var_qflx]
qflx_futArcSIC=xr.open_dataset(path+var_qflx+"_Amon_CESM1-WACCM-SC_"+exp2+"_combined.nc")[var_qflx]
qflx_pdSIC_2XCO2=xr.open_dataset(path+var_qflx+"_Amon_CESM1-WACCM-SC_"+exp4+"_combined.nc")[var_qflx]
qflx_futSIC_2XCO2=xr.open_dataset(path+var_qflx+"_Amon_CESM1-WACCM-SC_"+exp5+"_combined.nc")[var_qflx]

#convective precipitation
prc_pdSIC=xr.open_dataset(path+var_prc+"_Amon_CESM1-WACCM-SC_"+exp1+"_combined.nc")[var_prc]
prc_futArcSIC=xr.open_dataset(path+var_prc+"_Amon_CESM1-WACCM-SC_"+exp2+"_combined.nc")[var_prc]
prc_pdSIC_2XCO2=xr.open_dataset(path+var_prc+"_Amon_CESM1-WACCM-SC_"+exp4+"_combined.nc")[var_prc]
prc_futSIC_2XCO2=xr.open_dataset(path+var_prc+"_Amon_CESM1-WACCM-SC_"+exp5+"_combined.nc")[var_prc]

#transport (P-E)
rho=1000
transport_pdSIC = pr_pdSIC - qflx_pdSIC/rho
transport_futArcSIC = pr_futArcSIC - qflx_futArcSIC/rho
transport_pdSIC_2XCO2 = pr_pdSIC_2XCO2 - qflx_pdSIC_2XCO2/rho
transport_futSIC_2XCO2 = pr_futSIC_2XCO2 - qflx_futSIC_2XCO2/rho
```

```python
# 50years climatology
def climo (data):
    return data.fillna(0).mean("time")

#precipitation
pr_pdSIC_climo=climo(pr_pdSIC)
pr_pdSIC_2XCO2_climo=climo(pr_pdSIC_2XCO2)
pr_futArcSIC_climo=climo(pr_futArcSIC)
pr_futSIC_2XCO2_climo=climo(pr_futSIC_2XCO2)

#evaporation
qflx_pdSIC_climo=climo(qflx_pdSIC)
qflx_pdSIC_2XCO2_climo=climo(qflx_pdSIC_2XCO2)
qflx_futArcSIC_climo=climo(qflx_futArcSIC)
qflx_futSIC_2XCO2_climo=climo(qflx_futSIC_2XCO2)

#transport
transport_pdSIC_climo=climo(transport_pdSIC)
transport_pdSIC_2XCO2_climo=climo(transport_pdSIC_2XCO2)
transport_futArcSIC_climo=climo(transport_futArcSIC)
transport_futSIC_2XCO2_climo=climo(transport_futSIC_2XCO2)
```

```python
##
lat = pr_pdSIC_climo.lat
lon = pr_pdSIC_climo.lon
```

```python
#Difference
#precipitation
pr_futArcSIC_pdSIC_diff = pr_futArcSIC_climo - pr_pdSIC_climo
pr_futSIC_2XCO2_pdSIC_diff = pr_futSIC_2XCO2_climo - pr_pdSIC_climo

#evaporation
qflx_futArcSIC_pdSIC_diff = qflx_futArcSIC_climo - qflx_pdSIC_climo
qflx_futSIC_2XCO2_pdSIC_diff = qflx_futSIC_2XCO2_climo - qflx_pdSIC_climo

#transport
transport_futArcSIC_pdSIC_diff = transport_futArcSIC_climo - transport_pdSIC_climo
transport_futSIC_2XCO2_pdSIC_diff = transport_futSIC_2XCO2_climo - transport_pdSIC_climo
```
