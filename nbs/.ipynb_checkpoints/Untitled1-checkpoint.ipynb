```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

from matplotlib.colors import TwoSlopeNorm
norm = TwoSlopeNorm(vcenter = 0)
```

```python
# Define a function to group by season and compute the mean
def group_by_season_mean(ds, var_name):
    seasons = ds[var_name].groupby('time.season').mean(dim='time')
    return seasons

# Define a function to convert the units by multiplying by 1000
def convert_units_to_kg_m2_s(ds, var_name):
    ds[var_name] = ds[var_name] * 1000
    return ds

def calculate_transport(pr, qflx):
    return pr - qflx
```

```python
qflx1 = xr.open_dataset('/scratch/lvillanu/data/pa-pdSIC-ext_qflx_mean_each_month.nc')
qflx2 = xr.open_dataset('/scratch/lvillanu/data/pa-futArcSIC-ext_qflx_mean_each_month.nc')
qflx3 = xr.open_dataset('/scratch/lvillanu/data/pa-futAntSIC-ext_qflx_mean_each_month.nc')
qflx4 = xr.open_dataset('/scratch/lvillanu/data/pa-pdSIC-2XCO2-ext_qflx_mean_each_month.nc')
qflx5 = xr.open_dataset('/scratch/lvillanu/data/pa-futSIC-2XCO2-ext_qflx_mean_each_month.nc')

#qflx1_ = qflx1['qflx'].groupby('time.month').mean(dim='time')
#qflx2_ = qflx2['qflx'].groupby('time.month').mean(dim='time')
#qflx3_ = qflx3['qflx'].groupby('time.month').mean(dim='time')
#qflx4_ = qflx4['qflx'].groupby('time.month').mean(dim='time')
#qflx5_ = qflx5['qflx'].groupby('time.month').mean(dim='time')

# Group by season and compute mean for qflx
qflx1_seasonal = group_by_season_mean(qflx1, 'qflx')
qflx2_seasonal = group_by_season_mean(qflx2, 'qflx')
qflx3_seasonal = group_by_season_mean(qflx3, 'qflx')
qflx4_seasonal = group_by_season_mean(qflx4, 'qflx')
qflx5_seasonal = group_by_season_mean(qflx5, 'qflx')
```

```python
# Calculate transport (pr* - qflx*)
transport1 = calculate_transport(pr1_seasonal, qflx1_seasonal)
transport2 = calculate_transport(pr2_seasonal, qflx2_seasonal)
transport3 = calculate_transport(pr3_seasonal, qflx3_seasonal)
transport4 = calculate_transport(pr4_seasonal, qflx4_seasonal)
transport5 = calculate_transport(pr5_seasonal, qflx5_seasonal)
```

```python
qflx2 = xr.open_dataset('/scratch/lvillanu/data/pa-futArcSIC-ext_qflx_mean_each_month.nc')

qflx2_ = qflx2['qflx'].groupby('time.month').mean(dim='time')

# Assuming arctic_flux_mean is an xarray DataArray with dimensions (month, lat, lon)
# Calculate DJF mean (months 12, 1, 2)
djf_mean = qflx2_.sel(month=[12, 1, 2]).mean(dim='month')

# Calculate JJA mean (months 6, 7, 8)
jja_mean = qflx2_.sel(month=[6, 7, 8]).mean(dim='month')

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 8), subplot_kw={'projection': ccrs.NorthPolarStereo()})
seasons = {'DJF': djf_mean, 'JJA': jja_mean}

for ax, (season, data) in zip(axes, seasons.items()):
    ax.set_extent([-180, 180, 60, 90], crs=ccrs.PlateCarree())
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    
    # Plot with color
    c = ax.pcolormesh(data.lon, data.lat, data, transform=ccrs.PlateCarree(), cmap='OrRd')
    
    ax.set_title(f'{season} Mean')

# Add colorbar
cbar = fig.colorbar(c, ax=axes, fraction=0.02, pad=0.1)
cbar.set_label(f'Precipitation $m/s$')


fig.suptitle('Seasonal Mean Precipitation in the Arctic (50-year Mean)', y=1.02)
plt.show()
```
